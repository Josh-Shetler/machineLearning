```python
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow import keras

# create dataset from mnist digits
dataset = tf.keras.datasets.mnist

# split into train / test
(X_train, Y_train) , (X_test, Y_test) = dataset.load_data()

X_train, X_test = X_train/255, X_test/255

# manipulate data, to then create a validation set from X and Y train and Test
X = np.concatenate([X_train, X_test])
Y = np.concatenate([Y_train, Y_test])

datasetSize = 70000
trainRatio = 0.8
validationSize = 0.1
testRatio = 0.1

#creation of all train and test variables, including validation set X and Y
X_train, X_validation, Y_train, Y_validation = train_test_split(X, Y, test_size=(1-trainRatio))
X_validation, X_test, Y_validation, Y_test = train_test_split(X_validation, Y_validation, test_size=((testRatio/(validationSize+testRatio))))


#create a slice, and change datatype to make it manipulatable 
dataset = tf.data.Dataset.from_tensor_slices((X, Y))

# take first 80 percent of dataset
#train_dataset = dataset.take(int(trainRatio*datasetSize))

# create dataset using skip and take functionality. Skip over the size of the train dataset and take the remaining amount for validation. Next 10 percent
validation_dataset = dataset.skip(int(trainRatio*datasetSize)).take(int(validationSize*datasetSize))

# last 10 percent of data 
#test_dataset = dataset.skip(int(trainRatio*datasetSize)).skip(int(validationSize*datasetSize))

# initialize the tensorflow model 
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28,28)),
    tf.keras.layers.Dense(128, activation = 'relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(10, activation ='softmax')
])

#compile and optimize the model, define the loss function, and metrics involved.
model.compile(optimizer= 'adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

#train the model on 8 epochs 
model.fit(X_train, Y_train, epochs=1)

# get ending testLoss and testAccuracy of the model
testLoss, testAccuracy = model.evaluate(X_test, Y_test)

#list of 10,000 predictions on the model 
prediction = model.predict(X_test)

prediction[0]

#get ending valLoss and valAccuracy of the model
valLoss, valAccuracy = model.evaluate(X_validation, Y_validation)

round(valAccuracy, 5)
print("ending accuracy", valAccuracy)
```
